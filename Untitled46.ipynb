```python
# Step 1: Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', None)
sns.set(style='whitegrid')

#  Load the Dataset
df = pd.read_csv('netflix_titles.csv')

#  Explore Basic Info
print("Shape:", df.shape)
print("Info:")
print(df.info())
df.head()

#  Missing Values Check
print("\nMissing Values:")
print(df.isnull().sum())

#  Data Cleaning

df['country'] = df['country'].fillna('Unknown')
df['rating'] = df['rating'].fillna('Not Rated')

# Convert 'date_added' to datetime safely
df['date_added'] = pd.to_datetime(df['date_added'].str.strip(), errors='coerce')

# Extract year, month, day
df['year_added'] = df['date_added'].dt.year
df['month_added'] = df['date_added'].dt.month
df['day_added'] = df['date_added'].dt.day


# Drop rows with missing title or type (critical fields)
df.dropna(subset=['title', 'type'], inplace=True)

#Analysis & Questions

# Q1: Top 5 countries producing content
top_countries = df['country'].value_counts().head(5)
print("\nTop 5 Countries:\n", top_countries)

# Q2: Year with most releases
top_year = df['release_year'].value_counts().idxmax()
print("\nMost releases in:", top_year)

# Q3: Most common genres
df['listed_in'] = df['listed_in'].astype(str)
df_exploded = df.assign(genres=df['listed_in'].str.split(', ')).explode('genres')
top_genres = df_exploded['genres'].value_counts().head(10)
print("\nTop 10 Genres:\n", top_genres)

# Q4: Movies vs TV Shows
type_count = df['type'].value_counts()
print("\nType Count:\n", type_count)

# Q5: Top 5 Directors
top_directors = df['director'].dropna().value_counts().head(5)
print("\nTop 5 Directors:\n", top_directors)

# Q6: Indian Dramas
indian_dramas = df[
    df['country'].str.contains('India', na=False) &
    df['listed_in'].str.contains('Drama', na=False)
]
print("\nIndian Drama Titles:\n", indian_dramas[['title', 'release_year']].head())

# üìå Step 7: Visualizations

# Top 10 content-producing countries
plt.figure(figsize=(10,5))
top_countries.plot(kind='bar', color='skyblue')
plt.title('Top 5 Content Producing Countries')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Movies vs TV Shows Pie Chart
plt.figure(figsize=(6,6))
type_count.plot(kind='pie', autopct='%1.1f%%', colors=['#ff9999','#66b3ff'], startangle=140)
plt.title('Movies vs TV Shows')
plt.ylabel('')
plt.show()

# Releases per year line plot
plt.figure(figsize=(12,6))
df['release_year'].value_counts().sort_index().plot(kind='line', marker='o')
plt.title('Releases Per Year')
plt.xlabel('Year')
plt.ylabel('Number of Titles')
plt.grid(True)
plt.tight_layout()
plt.show()
```
